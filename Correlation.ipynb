```python
from sklearn import tree 
import pandas as pd
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join("Hollywood Attractiveness Attributes.csv"))
df.head()
```

```python
correlation = df.corr()
correlation = pd.DataFrame(correlation["Attractive"]).sort_values(
    by=["Attractive"], ascending=False).iloc[1:].style.format("{:.2%}")
# correlation.to_csv("Attractivness Correlation.csv")
correlation
```

```python
data = df.drop(["Image"], axis=1)
```

```python
corr = data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
# plt.savefig('Attractiveness Correlation')
plt.show()
```
